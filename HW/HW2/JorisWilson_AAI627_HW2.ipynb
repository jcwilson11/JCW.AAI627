## Setup / Load

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# load data
path = "EE627A_HW1_Data.csv"   
df = pd.read_csv(path)

# parse date and preview
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
print(df.shape)
print(df.columns.tolist())
df.head(10)
```

## Form Correlation Matrix

```python
# Full correlation matrix 
all_data = df.drop(columns=['Date']).copy()
corr_all = all_data.corr()

corr_all.to_csv("correlation_matrix_all.csv")

# print matrix 
corr_all.round(3)
```

## Factors

```python
factors = ['Mkt-RF', 'SMB', 'HML', 'Mom']
exclude = set(['Date', 'RF'] + factors) # eclude because not industry
industry_cols = [c for c in df.columns if c not in exclude]

factor_ind_corr = df[factors + industry_cols].corr().loc[factors, industry_cols]
factor_ind_corr.round(3)

# which facttor is strongest?
best_factor_per_industry = factor_ind_corr.abs().idxmax(axis=0)
best_corr_values = pd.Series(
    {ind: factor_ind_corr.loc[best_factor_per_industry[ind], ind] for ind in industry_cols},
    name="corr"
)
best_table = pd.DataFrame({"best_factor": best_factor_per_industry, "corr": best_corr_values}).loc[industry_cols]
best_table.sort_index().round(3)
```

```python
# Summary stats across industries (to answer "which factor correlates most highly with every industry?" overall) ---
avg_abs_corr_by_factor = factor_ind_corr.abs().mean(axis=1).sort_values(ascending=False)
min_corr_by_factor = factor_ind_corr.min(axis=1)
max_corr_by_factor = factor_ind_corr.max(axis=1)

print("Average correlation by factor (higher = stronger overall link):")
print(avg_abs_corr_by_factor.round(3))
print("\nMin correlation with industries (by factor):")
print(min_corr_by_factor.round(3))
print("\nMax correlation with industries (by factor):")
print(max_corr_by_factor.round(3))
```

```python
# check rf correlation
rf_corr = df[industry_cols].corrwith(df['RF']).sort_values()
rf_corr.round(3)
```

## Form the auto-correlation function (ACF)


```python
from statsmodels.tsa.stattools import acf

factors = ['Mkt-RF', 'SMB', 'HML', 'Mom']
n = len(df)
se = 1 / np.sqrt(n)               # standard error for white-noise ACF
upper = 1.96 * se
lower = -1.96 * se

acf_results = {}
for f in factors:
    series = df[f].dropna().values
    acf_vals = acf(series, nlags=10, fft=True)   # includes lag 0
    acf_results[f] = acf_vals

for f in factors:
    lags = np.arange(1, 11)
    vals = acf_results[f][1:11]
    plt.figure()
    plt.bar(lags, vals)
    plt.axhline(upper, linestyle='--')
    plt.axhline(lower, linestyle='--')
    plt.axhline(0)
    plt.title(f"ACF (lags 1–10): {f}")
    plt.xlabel("Lag")
    plt.ylabel("Autocorrelation")
    plt.show()
```
